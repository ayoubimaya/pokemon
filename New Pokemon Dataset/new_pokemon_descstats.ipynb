```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv('C://Users//mayoubi//Documents//Pokemon.csv', delimiter = ',')
pokemon.head()
```

```python
pokemon.info()
```

```python
pokemon["Type 2"].fillna(pokemon["Type 1"], inplace = True)
```

```python
pokemon.columns = pokemon.columns.str.upper().str.replace('_', '')
pokemon.head()
```

```python
pokemon["TYPE 1"].value_counts()
```

```python
pokemon["TYPE 2"].value_counts()
```

```python
pokemon.describe()
```

```python
pokemon = pokemon.drop(["#"],axis = 1)
pokemon.set_index("NAME")
```

```python
corr_matrix = pokemon.corr()
corr_matrix["HP"].sort_values(ascending = False)
```

```python
plt.hist(pokemon["ATTACK"])
plt.xlabel("Attack")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(pokemon["DEFENSE"])
plt.xlabel("Defense")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(pokemon["SPEED"])
plt.xlabel("Speed")
plt.ylabel("Count")
plt.show()
```

```python
fire_pokemon = pokemon[(pokemon["TYPE 1"] == 'Fire') | ((pokemon["TYPE 2"]) == "Fire")]
water_pokemon = pokemon[(pokemon["TYPE 1"] == 'Water') | ((pokemon["TYPE 2"]) == "Water")]
grass_pokemon = pokemon[(pokemon["TYPE 1"] == 'Grass') | ((pokemon["TYPE 2"]) == "Grass")]
plt.scatter(fire_pokemon.ATTACK.head(100), fire_pokemon.DEFENSE.head(100), color = 'R', label = "Fire", s = 40)
plt.scatter(water_pokemon.ATTACK.head(100), water_pokemon.DEFENSE.head(100), color = 'B', label = "Water", s = 40)
plt.scatter(grass_pokemon.ATTACK.head(100), grass_pokemon.DEFENSE.head(100), color = 'G', label = "Grass", s = 40)
plt.xlabel("Attack", fontsize = 18)
plt.ylabel("Defense", fontsize = 18)
plt.legend()
plt.plot()
fig=plt.gcf()
fig.set_size_inches(15,10)
plt.show()
```
